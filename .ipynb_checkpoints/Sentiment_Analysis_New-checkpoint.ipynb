```python
import pandas as pd
from textblob import TextBlob

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
file_path = "review_data.csv"

df = pd.read_csv(file_path, encoding="latin1")

print("Columns in dataset:")
print(df.columns)

df.head()
```

```python
review_column = "review"
true_label_column = "rating"
```

```python
def get_polarity(text):
    blob = TextBlob(str(text))
    return round(blob.sentiment.polarity, 2)
```

```python
df["polarity_score"] = df[review_column].apply(get_polarity)

df.head()
```

```python
def get_sentiment_label(score):
    
    if score > 0:
        return "positive"
    
    elif score < 0:
        return "negative"
    
    else:
        return "neutral"


df["predicted_sentiment"] = df["polarity_score"].apply(get_sentiment_label)

df.head()
```

```python
def convert_rating_to_sentiment(rating):

    if rating >= 4:
        return "positive"

    elif rating <= 2:
        return "negative"

    else:
        return "neutral"
```

```python
df["true_sentiment"] = df["rating"].apply(convert_rating_to_sentiment)
```

```python
y_true = df["true_sentiment"]
y_pred = df["predicted_sentiment"]
```

```python
accuracy = accuracy_score(y_true, y_pred)

precision = precision_score(y_true, y_pred, average="weighted")

recall = recall_score(y_true, y_pred, average="weighted")

f1 = f1_score(y_true, y_pred, average="weighted")

print("Model Performance Metrics\n")

print("Accuracy:", round(accuracy, 3))
print("Precision:", round(precision, 3))
print("Recall:", round(recall, 3))
print("F1 Score:", round(f1, 3))
```
